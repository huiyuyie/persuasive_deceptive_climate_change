```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eco = pd.read_csv('../data/economy-and-growth.csv')
evi = pd.read_csv('../data/environment.csv')
cli = pd.read_csv('../data/climate-change.csv')
ene = pd.read_csv('../data/energy-and-mining.csv')
```

```python
eco = eco.sort_values(by='Year', ascending=False).reset_index()
eco = eco.drop(columns='index')
eco
```

```python
evi = evi.sort_values(by='Year', ascending=False).reset_index()
evi = evi.drop(columns='index')
evi
```

```python
cli = cli.sort_values(by='Year', ascending=False).reset_index()
cli = cli.drop(columns='index')
cli
```

```python
ene = ene.sort_values(by='Year', ascending=False).reset_index()
ene = ene.drop(columns='index')
ene
```

```python
merged_df_1 = pd.merge(eco, evi, on=['Country Name', 'Country Code', 'Year'])
merged_df_2 = pd.merge(cli, ene, on=['Country Name', 'Country Code', 'Year'])
merged_df = pd.merge(merged_df_1, merged_df_2, on=['Country Name', 'Country Code', 'Year'])
merged_df = merged_df[merged_df['Year']>=2009]
```

#### Proposition: China with a larger population contribute more to CO₂ emissions and bear greater responsibility for climate change.

```python
target_emission = 'average_value_CO2 emissions (kt)'
target_pop = 'average_value_Population, total'

df = merged_df[
    merged_df['Country Name'].isin(['United States', 'China'])
].copy()

pivot = (
    df.pivot_table(
        index='Year',
        columns='Country Name',
        values=[target_emission, target_pop]
    )
    .reset_index()
    .sort_values('Year')
)

pivot.columns = [
    'Year',
    'China_emission', 'US_emission',
    'China_pop', 'US_pop'
]

x = np.arange(len(pivot['Year']))
width = 0.6

fig, axes = plt.subplots(1, 2, figsize=(14,6))

axes[0].bar(x, pivot['China_emission'], width, color='#d73027', alpha=0.7, label='China')
axes[0].bar(x, pivot['US_emission'], width, color='#4575b4', alpha=0.7, label='United States')

axes[0].set_title('CO₂ Emissions')
axes[0].set_xticks(x)
axes[0].set_xticklabels(pivot['Year'], rotation=45)
axes[0].set_ylabel('Total emissions')
axes[0].grid(axis='y', alpha=0.2)

axes[1].bar(x, pivot['China_pop'], width, color='#d73027', alpha=0.5, label='China')
axes[1].bar(x, pivot['US_pop'], width, color='#4575b4', alpha=0.5, label='United States')

axes[1].set_title('Population')
axes[1].set_xticks(x)
axes[1].set_xticklabels(pivot['Year'], rotation=45)
axes[1].set_ylabel('Total population')
axes[1].grid(axis='y', alpha=0.2)

fig.suptitle(
    'China emits more carbon than the United States from 2009 to 2018.\n'
    'This difference aligns with its larger population.',
    fontsize=14,
    y=1.02
)

axes[0].legend(frameon=False)

plt.tight_layout()
plt.show()
```

```python
target_pc = 'average_value_CO2 emissions (metric tons per capita)'

df_pc = merged_df[
    merged_df['Country Name'].isin(['United States', 'China'])
].copy()

pivot_pc = (
    df_pc
    .pivot_table(index='Year', columns='Country Name', values=target_pc)
    .reset_index()
    .sort_values('Year')
)

x = np.arange(len(pivot_pc['Year']))
width = 0.6

plt.figure(figsize=(12,6))

plt.bar(
    x,
    pivot_pc['United States'],
    width,
    color='#4575b4',
    alpha=0.7,
    label='United States',
    edgecolor='black'
)

plt.bar(
    x,
    pivot_pc['China'],
    width,
    color='#d73027',
    alpha=0.5,
    label='China',
    edgecolor='black'
)

plt.xticks(x, pivot_pc['Year'], rotation=45)
plt.ylabel('Carbon emissions per person')

plt.title(
    'The United States emits more carbon than China.\n'
    'Its emissions place a heavier burden on the environment.'
)

plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
