```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eco = pd.read_csv('../data/economy-and-growth.csv')
evi = pd.read_csv('../data/environment.csv')
cli = pd.read_csv('../data/climate-change.csv')
ene = pd.read_csv('../data/energy-and-mining.csv')
```

```python
eco = eco.sort_values(by='Year', ascending=False).reset_index()
eco = eco.drop(columns='index')
eco
```

```python
evi = evi.sort_values(by='Year', ascending=False).reset_index()
evi = evi.drop(columns='index')
evi
```

```python
cli = cli.sort_values(by='Year', ascending=False).reset_index()
cli = cli.drop(columns='index')
cli
```

```python
ene = ene.sort_values(by='Year', ascending=False).reset_index()
ene = ene.drop(columns='index')
ene
```

```python
merged_df_1 = pd.merge(eco, evi, on=['Country Name', 'Country Code', 'Year'])
merged_df_2 = pd.merge(cli, ene, on=['Country Name', 'Country Code', 'Year'])
merged_df = pd.merge(merged_df_1, merged_df_2, on=['Country Name', 'Country Code', 'Year'])
merged_df = merged_df[merged_df['Year']>=2010]
```

#### Proposition: China with a larger population contribute more to CO₂ emissions and bear greater responsibility for climate change.

```python
import matplotlib.pyplot as plt
import numpy as np

target_total = 'average_value_CO2 emissions (kt)'

df = merged_df[
    (merged_df['Country Name'].isin(['China', 'United States'])) &
    (merged_df['Year'].between(2009, 2018))
].copy()

# world total
world = merged_df[
    (merged_df['Country Name'] == 'World') &
    (merged_df['Year'].between(2009, 2018))
][['Year', target_total]].rename(columns={target_total: 'world_total'})

df = df.merge(world, on='Year')

# transformation
df['share'] = df[target_total] / df['world_total'] * 100

pivot = df.pivot_table(
    index='Year',
    columns='Country Name',
    values='share'
).reset_index()

x = np.arange(len(pivot['Year']))
width = 0.38

fig, ax = plt.subplots(figsize=(12,8))

# bars
ax.bar(
    x - width/2,
    pivot['China'],
    width,
    color='#d73027',
    label='China'
)

ax.bar(
    x + width/2,
    pivot['United States'],
    width,
    color='#4575b4',
    label='United States'
)

ax.set_title(
    'The Cost of Catching Up:\nChina Drives More Warming',
    loc='left',
    fontsize=30,
    fontweight='bold',
    fontfamily='Georgia',
    pad=30
)

ax.text(
    0, 1.00,
    'China consistently accounts for a larger share of global carbon emissions than the United States.',
    transform=ax.transAxes,
    ha='left',
    va='bottom',
    fontsize=17,
    fontfamily='Georgia'
)

# axis
ax.set_ylabel(
    'Share of global CO₂ emissions (%)',
    fontsize=13,
    fontweight='bold'
)

ax.set_xlabel(
    'Year',
    fontsize=13,
    fontweight='bold'
)

ax.set_xticks(x)
ax.set_xticklabels(pivot['Year'], rotation=45, fontsize=10)
ax.tick_params(axis='y', labelsize=10)

ax.text(
    0, -0.14,
    'Global Share = country CO₂ emissions (kt) ÷ world CO₂ emissions (kt) × 100',
    transform=ax.transAxes,
    ha='left',
    va='top',
    fontsize=12,
    color='dimgray',
    fontfamily='Georgia'
)

ax.text(
    1, -0.14,
    'Source: Human Development Indicators, 1960–2020 (World Bank)',
    transform=ax.transAxes,
    ha='right',
    va='top',
    fontsize=10,
    color='dimgray',
    fontfamily='Georgia'
)

# style
ax.grid(axis='y', alpha=0.2)
ax.legend(frameon=False)

plt.subplots_adjust(top=0.85)
plt.tight_layout()
plt.show()
```

```python
target_pc = 'average_value_CO2 emissions (metric tons per capita)'

df_pc = merged_df[
    (merged_df['Country Name'].isin(['China', 'United States'])) &
    (merged_df['Year'].between(2009, 2018))
].copy()

pivot_pc = df_pc.pivot_table(
    index='Year',
    columns='Country Name',
    values=target_pc
).reset_index()

x = np.arange(len(pivot_pc['Year']))
width = 0.38

fig, ax = plt.subplots(figsize=(12,8))

ax.bar(
    x - width/2,
    pivot_pc['China'],
    width,
    color='#d73027',
    label='China'
)

ax.bar(
    x + width/2,
    pivot_pc['United States'],
    width,
    color='#4575b4',
    label='United States'
)

ax.set_title(
    'Who’s Really Driving Global Warming?\nThe U.S. Leaves a Larger Footprint.',
    loc='left',
    fontsize=25,
    fontweight='bold',
    fontfamily='serif',
    pad=30
)

ax.text(
    0, 1.00,
    'The United States consistently has higher carbon emissions per person than China.',
    transform=ax.transAxes,
    ha='left',
    va='bottom',
    fontsize=17,
    fontfamily='serif'
)

ax.set_ylabel(
    'CO₂ emissions per capita (metric tons)',
    fontsize=13,
    fontweight='bold'
)

ax.set_xlabel(
    'Year',
    fontsize=13,
    fontweight='bold'
)

ax.set_xticks(x)
ax.set_xticklabels(pivot_pc['Year'], rotation=45)

ax.text(
    0, -0.14,
    'CO₂ per capita (metric tons) = total CO₂ emissions (kt) × 1000 ÷ population',
    transform=ax.transAxes,
    ha='left',
    va='top',
    fontsize=12,
    color='dimgray',
    fontfamily='Georgia'
)

ax.text(
    1, -0.14,
    'Source: Human Development Indicators, 1960–2020 (World Bank)',
    transform=ax.transAxes,
    ha='right',
    va='top',
    fontsize=10,
    color='dimgray',
    fontfamily='Georgia'
)

ax.grid(axis='y', alpha=0.2)
ax.legend(frameon=False)

plt.tight_layout()
plt.show()
```
